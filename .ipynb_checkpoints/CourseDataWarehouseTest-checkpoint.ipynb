```python
%%sql
SELECT 
    (SELECT Count(*) FROM COURSE1) as CatalogCourses,
    (SELECT SUM(Meeting_times) FROM CLASS_STATISTICS) as CourseMeetings;
```

```python
%%sql
SELECT Term, Program_name, Location, Name, CatalogID, Course_title, Meeting_times, Meeting_hours
 FROM CLASS_STATISTICS
    LEFT JOIN TERM1 USING (Term_id)
    LEFT JOIN PROGRAM1 USING (Program_id)
    LEFT JOIN LOCATION1 USING (Location_id)
    LEFT JOIN INSTRUCTOR1 USING (Instructor_id)
    LEFT JOIN COURSE1 USING (Course_id)
    WHERE Name like '%Huntley';
```
