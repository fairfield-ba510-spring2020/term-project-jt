## **Query 1**

### The total number of hours each professor taught in **School Year 2018**

```python
%%sql

SELECT Name, Term, SUM(Meeting_hours)
    FROM CLASS_STATISTICS
        LEFT JOIN INSTRUCTOR1 USING (Instructor_id)
        LEFT JOIN TERM1 USING (Term_id)
        WHERE Term like '%2018'
  GROUP BY Term, Name
LIMIT 5
```

## **Query 2**

### The total number of professors within each program for **School Year 2018**

```python
%%sql

SELECT Program_name, COUNT(DISTINCT Instructor_id) as Number_of_Professors
    FROM CLASS_STATISTICS
        LEFT JOIN PROGRAM1 USING (Program_id)
        LEFT JOIN TERM1 USING (Term_id)
        WHERE Term like '%2018'
        GROUP BY Program_name
        ORDER BY Number_of_Professors DESC
    LIMIT 5
```

## **Query 3**

### Most used classrooms within a building in **Year 2018**

#### Example: DSB

```python
%%sql
SELECT Location, Term, COUNT(ClassStat_id) as Total_Number_of_Classes
FROM CLASS_STATISTICS
    LEFT JOIN LOCATION1 USING (Location_id)
    LEFT JOIN TERM1 USING (Term_id)
    WHERE Location like 'DSB%'
    GROUP BY Location, Term
    LIMIT 5
```

### The student-professor ratio by programs

```python
%%sql
SELECT Term, Program_name as Program, SUM(Act) as Seats, AVG(Act) as Student_Professor_Ratio, COUNT(*) as Sections
FROM CLASS_STATISTICS
    LEFT JOIN TERM1 USING (Term_id)
    LEFT JOIN PROGRAM1 USING (Program_id)
    WHERE Program_name IN ('Accounting','Economics','Finance','Management','Marketing','Information Systems',
                           'Operations Management','Business')
    GROUP BY Program_name, Term
    
```
